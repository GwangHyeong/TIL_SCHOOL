## 다단계 피벗 테이블(pivot table)
\- pd.cut 구문사용: pd.cut(a, b) 함수는 인자(카테고리할 숫자 data, 잘라낼 구분 값)을 넣어 쉽게 category화 할 수 있음<br>
\- pd.cut(a, b)로 구분된 dataset들은 category datatype series로 반환 됨

```python
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table('survived', ['sex', age], 'class')
```

```python
age = pd.cut(titanic['age'], [0,10,20,30,40,50,60,70])
titanic.pivot_table('survived', ['sex', age], 'class')
```

\- column기준으로 동작할 때도 같은 방법을 사용

```python
fare = pd.qcut(titanic['fare'],2)
titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
```

## 기타 pivot table option

###### (2) 기타 pivot table option
\- DataFrame의 pivot_table method의 전체 호출 시그니처는 다음과 같음

DataFrame.pivot_table(data, values=None, index=None, columns=None,
                     aggfunc='mean', fill_value=None, margins=False,
                     dropna=True, margins_name='All')

```python
titanic.pivot_table(index='sex', columns='class',
                   aggfunc={'survived':sum, 'fare':'mean'})
```

```python
titanic.pivot_table('survived', index='sex', columns='class', margins=True)
```

margins=True의 의미: 모든 데이터를 분석한 결과를 오른쪽과 아래에 붙일지 여부
